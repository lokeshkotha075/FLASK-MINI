```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

```python
##data cleaning
```

```python
##missing values
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converterd into tow sets based on region from 122th index,we can make a new ccolum bassed
on the region

add new column with region


```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df=dataset
```

```python
df.info()
```

```python
df["Region"]=df[["Region"]].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
##remove the null value index
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
##remove the 122 row
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
##fix spaces in columsn names
df.columns=df.columns.str.strip()
```

```python
df.columns
```

##changes the required colums as interger datra type

```python
df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
 ###changing the other columsns to float datatyoe
```

```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
        
```

```python
df.info()
```

```python
df.head()
```

```python
##let save the cleaned dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```
